```python
G = nx.Graph()

with open('../data/MERGED.json', 'r') as f1:
        DATA = json.load(f1)
        
with open('../data/ROOTS.json', 'r')as f2:
        ROOTS = json.load(f2)

for item in DATA:
    try:
        item_id = item['fd_id']
        
        try:
            item_title = item['mms_title']
        except:
            item_title = "NA"
        try:
            item_code = item['mms_code']
        except:
            item_code = "NA"
        try:
            item_defn = item['fd_defn']
        except:
            item_defn = "NA"
        try:
            item_childs = item['fd_childs']
        except:
            item_childs = []
        
        G.add_node(item_id,\
                   title = item_title,\
                   code = item_code,\
                   defn = item_defn,\
                   childs = item_childs)
        
        
        childs = item_childs
        if len(childs)>0:
            for c_id in childs:
                G.add_edge(item_id,c_id)
    except:
        pass
            
            
G.add_node("ICD11",\
           title="International Disease CLassification",\
           defn = "International Disease CLassification",\
           code = "ICD11")

for root in ROOTS:
        G.add_edge("ICD11",root)
```

```python
nx.info(G)
```

```python
nx.is_connected(G)
```

```python
ROOT = [n for n in G.neighbors("ICD11")]
    
INFO = []
for node in ROOT:
       
        try:
            INFO.append({"ID": node,\
                     "title": G.nodes()[node]["title"],\
                     "defn":  G.nodes()[node]['defn'],\
                     "child": len([n for n in G.neighbors(node)])})
        except:
            
             INFO.append({"ID": node,\
                     "title": "NA",\
                     "defn":"NA",\
                     "child": "NA"})
```
